```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("/content/combined.xlsx")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
df.isnull().sum()
df1 = df.dropna()
df1.isnull().sum()
df1.info()
```

```python
df1=df1.drop(labels='entry_id', axis=1)
```

```python
df1=df1.drop(labels='created_at', axis=1)
```

```python
df1.corr()
```

```python
sns.pairplot(df1)
```

```python
x=df1.iloc[:,:-1]
y=df1.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression= LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
print(regression.coef_)
```

```python
reg_pred=regression.predict(x_test)
```

```python
print(reg_pred)
```

```python
plt.scatter(reg_pred,y_test)
```

```python
residuals=y_test-reg_pred
```

```python
residuals
```

```python
sns.displot(residuals,kind='kde')
```

```python
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))

#for root mean squre error
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

```python
1-((1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1))
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()
prams={'alpha':[1e-15,1e-10,1e-7,1e-5,1e-4,1e-2,1e-1,1,5,10,30,100,155,300,10000]}
ridge_reg=GridSearchCV(ridge,prams,scoring='neg_mean_squared_error',cv=5)
ridge_reg.fit(x,y)
```

```python
print(ridge_reg.best_score_)
print(ridge_reg.best_params_)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lasso=Lasso()
prams={'alpha':[1e-15,1e-10,1e-7,1e-5,1e-4,3343e-4,1e-1,1,5,10,30,100,155,300,1000]}
lasso_reg=GridSearchCV(lasso,prams,scoring='neg_mean_squared_error',cv=5)
lasso_reg.fit(x,y)
```

```python
print(lasso_reg.best_score_)
print(lasso_reg.best_params_)
```

```python
pred_lasso=lasso_reg.predict(x_test)
pred_ridge=ridge_reg.predict(x_test)
```

```python
sns.distplot(y_test-pred_ridge)
```

```python
sns.distplot(y_test-pred_lasso)
```
